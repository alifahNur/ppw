```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

# Kalimat
kalimat = "Saya makan nasi"

# Tokenisasi
tokenizer = Tokenizer()
tokenizer.fit_on_texts([kalimat])
total_kata = len(tokenizer.word_index) + 1

# Mengonversi kalimat menjadi urutan token
sekuens = tokenizer.texts_to_sequences([kalimat])[0]

# One-hot encoding untuk input dan output
input_sekuens = []
output_sekuens = []
for i in range(1, len(sekuens)):
    input_sekuens.append(sekuens[:i])
    output_sekuens.append(sekuens[i])

# Padding input sequence ke panjang yang sama
max_len = max([len(x) for x in input_sekuens])
input_sekuens = np.array([np.pad(x, (0, max_len-len(x))) for x in input_sekuens])

# One-hot encoding
input_sekuens = to_categorical(input_sekuens, num_classes=total_kata)
output_sekuens = to_categorical(output_sekuens, num_classes=total_kata)

# Definisikan model
model = Sequential()
model.add(Flatten(input_shape=(input_sekuens.shape[1], total_kata)))  # Flatten dimensi input
model.add(Dense(64, activation='relu'))
model.add(Dense(total_kata, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Latih model (iterasi sebanyak 2 kali)
model.fit(input_sekuens, output_sekuens, epochs=2)
```
