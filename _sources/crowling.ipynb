```python
from bs4 import BeautifulSoup
import requests
```

```python
from datetime import datetime, timedelta
```

```python
url ='https://www.liputan6.com/news'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Temukan elemen <div> dengan kelas 'overflow margin-bottom-xs'
judul = soup.find('div', class_= "articles--iridescent-list")

# Jika elemen ditemukan, cari semua elemen <span> di dalamnya dan tampilkan teksnya
if judul:
    # Inisialisasi counter
    counter = 1

    for span in judul.find_all('span', class_= "articles--iridescent-list--text-item__title-link-text"):
        print(f"{counter}. {span.text}")
        counter += 1
else:
    print("Elemen dengan kelas 'overflow margin-bottom-xs' tidak ditemukan.")
```

```python
link_counter = 1
print("\nDaftar Link:")
seen_links = set()  # Set untuk melacak link yang sudah dicetak
for link in judul.find_all('a'):
  href = link.get('href')
  if href and href not in seen_links:  # Periksa apakah href ada dan belum dicetak
    print(f"{link_counter}. {href}")
    seen_links.add(href)  # Tambahkan href ke set
    link_counter += 1
```

```python

# Temukan semua artikel dengan kelas 'text-card'
articles = soup.find_all('article', class_='text-card')

# Inisialisasi counter untuk artikel
article_counter = 1

# Iterasi melalui setiap artikel
for article in articles:
    # Ambil judul dan link
    title = article.find('h2', class_='title')
    if title:
        title_text = title.text.strip()
        link = title.find('a')['href']
        print(f"{article_counter}. Judul: {title_text}")
        print(f"   Link: {link}")

    # Ambil isi artikel
    content = article.find('p')
    if content:
        print(f"   Isi: {content.text.strip()}")

    # Ambil tanggal publikasi
    date = article.find('h4', class_='date')
    if date:
        date_text = date.text.strip()

        # Konversi "X jam lalu" ke waktu spesifik
        if "jam lalu" in date_text:
            hours_ago = int(date_text.split(' ')[0])
            publish_time = datetime.now() - timedelta(hours=hours_ago)
            print(f"   Tanggal: {publish_time.strftime('%d-%m-%Y %H:%M')}")
        else:
            print(f"   Tanggal: {date_text}")

    # Increment counter
    article_counter += 1
```
