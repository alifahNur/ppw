```python
from bs4 import BeautifulSoup
import requests
```

```python
from datetime import datetime, timedelta
```

```python
url ='https://www.liputan6.com/news'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Temukan elemen <div> dengan kelas 'overflow margin-bottom-xs'
judul = soup.find('div', class_='articles')

# Jika elemen ditemukan, cari semua elemen <span> di dalamnya dan tampilkan teksnya
if judul:
    # Inisialisasi counter
    counter = 1

    for span in judul.find_all('span', class_= "articles--iridescent-list--text-item__title-link-text"):
        print(f"{counter}. {span.text}")
        counter += 1
else:
    print("Elemen dengan kelas 'overflow margin-bottom-xs' tidak ditemukan.")
```

```python
link_counter = 1
print("\nDaftar Link:")
seen_links = set()  # Set untuk melacak link yang sudah dicetak
for link in judul.find_all('a',class_="ui--a articles--iridescent-list--text-item__title-link"):
  href = link.get('href')
  if href and href not in seen_links:  # Periksa apakah href ada dan belum dicetak
    print(f"{link_counter}. {href}")
    seen_links.add(href)  # Tambahkan href ke set
    link_counter += 1
```

```python
# Mengambil semua artikel dari halaman
articles = soup.find_all('article')  # Ubah selector menjadi 'article' jika setiap artikel dibungkus dalam elemen <article>

# Set untuk menyimpan href yang sudah dilihat agar tidak terjadi duplikasi
seen_links = set()

# Counter untuk artikel yang diproses
link_counter = 0

# Iterasi melalui setiap artikel
for article in articles:
    # Ambil link artikel
    link = article.find('a', class_='ui--a articles--iridescent-list--text-item__title-link')
    if link:
        href = link.get('href')
        title = link.get('title').strip()

        if href and href not in seen_links:
            print(f"Artikel {link_counter + 1}:")
            print(f"   Judul: {title}")
            print(f"   Link: {href}")

            # Ambil tanggal publikasi
            date = article.find('time', class_='articles--iridescent-list--text-item__time')
            if date:
                date_text = date.text.strip()
                print(f"   Tanggal: {date_text}")

            # Ambil kategori artikel
            category = article.find('a', class_='articles--iridescent-list--text-item__category')
            if category:
                category_text = category.text.strip()
                print(f"   Kategori: {category_text}")

            # Mencari dan mencetak ringkasan artikel yang sesuai
            summary_div = article.find('div', class_='articles--iridescent-list--text-item__summary articles--iridescent-list--text-item__summary-seo')
            if summary_div:
                summary_text = summary_div.text.strip()
                print(f"   Ringkasan: {summary_text}")

            # Tambahkan href ke set untuk menghindari duplikasi
            seen_links.add(href)
            link_counter += 1
            print()  # Spasi untuk memisahkan artikel

# Menampilkan total artikel yang ditemukan
print(f"Total artikel yang ditampilkan: {link_counter}")
```
